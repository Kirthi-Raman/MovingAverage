```python
# Download the packages
import pandas as pd
import matplotlib.pyplot as plt
import requests
```

```python
company = "AAPL"
```

```python
# Extract the stock price data from Website using API
stockprice = requests.get(f"https://financialmodelingprep.com/api/v3/historical-price-full/AAPL?apikey=demo")
```

```python
stockprice
```

```python
stockprice = stockprice.json()
stockprice
```

```python
stockprice = stockprice['historical'][:]
```

```python
stockprice
```

## Convert the stock data from dictionary format to Pandas Dataframe

```python
stock_df = pd.DataFrame.from_dict(stockprice)
```

```python
stock_df.head()
```

```python
# Reverse the row values: set the dataframe in the ascending order oof dates
stock_df = stock_df.iloc[::-1]
```

```python
stock_df
```

```python
# Set 'date' as index
stock_df = stock_df.set_index('date')
stock_df.head()
```

## Determine the Short Term and Long Term MA

```python
# Short Term MA
stock_df['20d'] = stock_df['close'].rolling(20).mean()
stock_df['20d']
```

```python
# Long Term MA
stock_df['250d'] = stock_df['close'].rolling(250).mean()
stock_df['250d']
```

```python
# Plot the MA
stock_df[['close','20d','250d']].plot(figsize=(10,4))
plt.grid(True)
plt.title(company + ' Moving Average')
plt.axis('tight')
plt.ylabel('Price')
```
